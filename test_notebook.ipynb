```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
home_path=os.path.expanduser('~')
file=pd.read_csv(os.path.join(home_path,'TV_Unwrapped', 'TV Wrapped - Sheet1.csv'))
```

```python
params=['Genre', 'Platform', 'Rewatch', 'Finish Date']
for param in params:
    df = file.groupby(param)[param].count()
    plt.figure(figsize=(5,5))
    plot = df.plot.pie(y=param)
    plt.show()
```

```python
params=['Genre', 'Platform', 'Rewatch', 'Finish Date']
for param in params:
    df = file.groupby(param)['Runtime'].sum()
    plt.figure(figsize=(5,5))
    plot = df.plot.pie(y=param)
    plt.show()
```

## Try plotly instead for nicer plots

```python
import plotly.express as px
import plotly.io as pio
pio.renderers.default = 'iframe' # or 'notebook' or 'colab' or 'jupyterlab'
params=['Genre', 'Platform', 'Rewatch', 'Finish Date']
for param in params:
    df = file.groupby(param)['Runtime'].sum().reset_index()
    fig = px.pie(df, values='Runtime', names=param)
    fig.show()
```
